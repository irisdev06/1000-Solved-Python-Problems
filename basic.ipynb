### áƒ¦â˜…áƒ¦ 9. Reverse a String with Slicing áƒ¦â˜…áƒ¦
Slicing syntax is:

text[start : end : step]

- start â†’ where to begin
- end â†’ where to stop
- step â†’ how many positions to move each time

So this code does the following:

- start â†’ omitted â†’ start from the beginning
- end â†’ omitted â†’ go until the end
- step = -1 â†’ move backwards



```python
text = "Iris"
reversed_text = text[::-1]
print(reversed_text)
```
